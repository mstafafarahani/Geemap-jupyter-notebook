## Let's Show the map

```python
import ee
import geemap
import datetime


# Create an interactive map
Map = geemap.Map(center=[0, 0], zoom=2)

# Define a geometry manually or use a country for simplicity
# Example: Use Kenya's boundary
country = ee.FeatureCollection("FAO/GAUL/2015/level0").filter(ee.Filter.eq('ADM0_NAME', 'Kenya'))

# Center map
Map.centerObject(country, 6)
Map.addLayer(country, {}, "Country")

# Load MODIS land cover data
forest_ic = ee.ImageCollection("MODIS/061/MCD12Q1") \
    .select('LC_Type1') \
    .filterBounds(country) \
    .filterDate('2001-01-01', '2023-01-01')

print('Image count:', forest_ic.size().getInfo())

# Function to calculate forest area (value 1: Evergreen Needleleaf Forests)
def calculate_forest_area(image):
    clipped = image.clip(country)
    forest = clipped.eq(1)  # Value 1 for Evergreen Needleleaf Forests
    masked_forest = forest.updateMask(forest)
    forest_area_km2 = masked_forest.multiply(ee.Image.pixelArea().divide(1e6))
    return forest_area_km2.set({
        'system:time_start': image.get('system:time_start'),
        'system:time_end': image.get('system:time_end')
    })

forest_areas = forest_ic.map(calculate_forest_area)



# Display forest cover sum layer
forest_sum = forest_areas.sum()
Map.addLayer(forest_sum, {'min': 32, 'max': 511, 'palette': ['1aff00']}, 'Forest Area Growth')

# Add a legend
legend_dict = {'Forest Growth': '1aff00'}
Map.add_legend(title="Forest Area Growth", legend_dict=legend_dict)

# Show the map
Map
```

```python
print('Image count:', forest_ic.size().getInfo())
```

```python
# Chart forest area over time
chart = geemap.chart.image_series(
    image_collection=forest_areas,
    region=country.geometry(),
    reducer=ee.Reducer.sum(),
    scale=500,
    series_property='system:time_start',
    title='Forest Area Growth (km)',
    ylabel='Forest Area (km)'
)
chart
```

```python
import ee
import pandas as pd
from google.colab import files



# Load country boundary (change 'Kenya' to your target)
country = ee.FeatureCollection("FAO/GAUL/2015/level0") \
            .filter(ee.Filter.eq('ADM0_NAME', 'Kenya'))

# Load MODIS forest land cover
modis = ee.ImageCollection("MODIS/061/MCD12Q1") \
    .select("LC_Type1") \
    .filterBounds(country) \
    .filterDate("2001-01-01", "2023-01-01")

# Function to compute forest area (class 1 = Evergreen Needleleaf Forest)
def compute_area(image):
    year = ee.Date(image.get("system:time_start")).format("YYYY")
    forest_mask = image.eq(1)
    forest_area = forest_mask.multiply(ee.Image.pixelArea()).divide(1e6)
    stats = forest_area.reduceRegion(
        reducer=ee.Reducer.sum(),
        geometry=country.geometry(),
        scale=500,
        maxPixels=1e13
    )
    return ee.Feature(None, {
        "year": year,
        "forest_area_km2": stats.get("LC_Type1")
    })

# Map the function
features = ee.FeatureCollection(modis.map(compute_area))

# Get the results to Python
# This uses client-side evaluation with getInfo
data = features.aggregate_array('year').getInfo()
areas = features.aggregate_array('forest_area_km2').getInfo()

# Convert to DataFrame
df = pd.DataFrame({
    "year": data,
    "forest_area_km2": areas
})

# Save CSV
csv_filename = "forest_area_timeseries.csv"
df.to_csv(csv_filename, index=False)

# Trigger download to your computer
files.download(csv_filename)
```
